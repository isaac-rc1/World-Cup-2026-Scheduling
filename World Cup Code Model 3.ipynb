# StartUp

```julia
using JuMP
using Gurobi
```

# Constants and Coefficients

```julia
Stadiums = 1:16;
Countries = 1:48;
Days = 1:17;
Times = 1:3;
Matches = 1:72;

Hosts = [1 2 3]; #MX, US, CAN
Not8 = [1 2]; #12 PM, 3 PM

DayCount = 17;
ShortDay = 1:15;
StadiumCount = 16;
CountryCount = 48;
MatchCount = 72;

VD = [[0 140 800 1050 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 ];
    [140 0 680 960 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 ];
    [800 680 0 380 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 ];
    [1050 960 380 0 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 ];
    [1000000 1000000 1000000 1000000 0 570 340 930 940 1160 1330 1000000 1000000 1000000 1000000 1000000 ]; 
    [1000000 1000000 1000000 1000000 570 0 420 610 510 780 1020 1000000 1000000 1000000 1000000 1000000 ]; 
    [1000000 1000000 1000000 1000000 340 420 0 940 880 1050 1220 1000000 1000000 1000000 1000000 1000000 ]; 
    [1000000 1000000 1000000 1000000 930 610 940 0 240 460 780 1000000 1000000 1000000 1000000 1000000 ]; 
    [1000000 1000000 1000000 1000000 940 510 880 240 0 650 800 1000000 1000000 1000000 1000000 1000000 ]; 
    [1000000 1000000 1000000 1000000 1160 780 1050 460 650 0 800 1000000 1000000 1000000 1000000 1000000 ]; 
    [1000000 1000000 1000000 1000000 1330 1020 1220 780 800 800 0 1000000 1000000 1000000 1000000 1000000 ]; 
    [1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 0 550 490 510 1480];
    [1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 550 0 220 310 1250];
    [1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 490 220 0 100 1280];
    [1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 510 310 100 0 1210];
    [1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1000000 1480 1250 1280 1210 0]]

LD = [1 3 4 4 4 4 4 4 4 4 4 4 4 6 6 6 6];
```

# Model Start

```julia
model = Model(Gurobi.Optimizer);
```

# Decision Variables

```julia
# Binary: Country c plays 1st, 2nd, 3rd round at Stadium s
@variable(model, RA1[c1 in Countries, s in Stadiums], Bin)
@variable(model, RB1[c1 in Countries, s in Stadiums], Bin)
@variable(model, RC1[c1 in Countries, s in Stadiums], Bin)

@variable(model, RA2[c1 in Countries, s in Stadiums], Bin)
@variable(model, RB2[c1 in Countries, s in Stadiums], Bin)
@variable(model, RC2[c1 in Countries, s in Stadiums], Bin)

# Binary: Stadium s hosts match m
@variable(model, MS[s in Stadiums, m in Matches], Bin)

# Binary: Match m happens on day d at time t
@variable(model, MT[d in Days, t in Times, m in Matches], Bin)

# Binary: Country c plays match m
@variable(model, MC[c in Countries, m in Matches], Bin)

# Binary: Match m happens on day d
@variable(model, MD[d in Days, m in Matches], Bin)

# Binary: Country c travels from stadium s1 to s2 (after 1st and 2nd game)
@variable(model, VAB[c in Countries, s1 in Stadiums, s2 in Stadiums], Bin)

# Binary: Country c travels from stadium s2 to s3 (after 2nd and 3rd game)
@variable(model, VBC[c in Countries, s2 in Stadiums, s3 in Stadiums], Bin)
```

# Objective Function

```julia
@objective(model, Min, 
    sum(VAB[c, s1, s2] * VD[s1, s2] for c in Countries, s1 in Stadiums, s2 in Stadiums) +
    sum(VBC[c, s2, s3] * VD[s2, s3] for c in Countries, s2 in Stadiums, s3 in Stadiums)
)


#backup objective
#@objective(model, Min, 
#    sum(VAB[c, s1, s2] * VD[s1, s2] for c in Countries, s1 in Stadiums, s2 in Stadiums if s1 != s2) +
#    sum(VBC[c, s2, s3] * VD[s2, s3] for c in Countries, s2 in Stadiums, s3 in Stadiums if s2 != s3)
#)
```

# Constraints

## Model Builders

### Round Robin

```julia
@constraint(model, [c in Countries], sum(MC[c, m] for m in Matches) == 3)
```

### Match Ensuring @ Stadium

```julia
@constraint(model, [c1 in Countries, s in Stadiums, m in Matches], MS[s, m] <= RA1[c1, s])
@constraint(model, [c2 in Countries, s in Stadiums, m in Matches], MS[s, m] <= RA2[c2, s])
@constraint(model, [c1 in Countries, c2 in Countries, s in Stadiums, m in Matches], MS[s, m] >= RA1[c1, s] + RA2[c2, s] - 1)

@constraint(model, [c1 in Countries, s in Stadiums, m in Matches], MS[s, m] <= RB1[c1, s])
@constraint(model, [c2 in Countries, s in Stadiums, m in Matches], MS[s, m] <= RB2[c2, s])
@constraint(model, [c1 in Countries, c2 in Countries, s in Stadiums, m in Matches], MS[s, m] >= RB1[c1, s] + RB2[c2, s] - 1)

@constraint(model, [c1 in Countries, s in Stadiums, m in Matches], MS[s, m] <= RC1[c1, s])
@constraint(model, [c2 in Countries, s in Stadiums, m in Matches], MS[s, m] <= RC2[c2, s])
@constraint(model, [c1 in Countries, c2 in Countries, s in Stadiums, m in Matches], MS[s, m] >= RC1[c1, s] + RC2[c2, s] - 1)

```

### Match Occurs in Only one stadium

```julia
@constraint(model, [m in Matches], sum(MS[s, m] for s in Stadiums) == 1)
```

### Travel Logic

```julia
@constraint(model, [c in Countries, s1 in Stadiums, s2 in Stadiums], 
    VAB[c, s1, s2] >= RA1[c, s1] + RB1[c, s2] - 1)
@constraint(model, [c in Countries, s1 in Stadiums, s2 in Stadiums], 
    VAB[c, s1, s2] <= RA1[c, s1])
@constraint(model, [c in Countries, s1 in Stadiums, s2 in Stadiums], 
    VAB[c, s1, s2] <= RB1[c, s2])


@constraint(model, [c in Countries, s1 in Stadiums, s2 in Stadiums], 
    VBC[c, s1, s2] >= RA1[c, s1] + RB1[c, s2] - 1)
@constraint(model, [c in Countries, s1 in Stadiums, s2 in Stadiums], 
    VBC[c, s1, s2] <= RA1[c, s1])
@constraint(model, [c in Countries, s1 in Stadiums, s2 in Stadiums], 
    VBC[c, s1, s2] <= RB1[c, s2])

@constraint(model, [c in Countries, s in Stadiums], 
    RA1[c, s] <= sum(MS[s, m] * MC[c, m] for m in Matches)) # rough sketch
```

### Match is between two countries

```julia
@constraint(model, [m in Matches], sum(MC[c, m] for c in Countries) == 2)
```

### Match Ensuring @ Time

```julia
@constraint(model, [d in Days, m in Matches], sum(MT[d, t, m] for t in Times) == MD[d, m])

@constraint(model, [m in Matches], sum(MD[d, m] for d in Days) == 1)
```

## Host Nation Constraints

### Host Nations Only pPlay @ 8 PM

```julia
@constraint(model,[m in Matches, c in Countries],
            sum(MT[d, t, m] for d in Days, t in Not8) + MC[c, m] <= 1
        )
```

### Mexico Opening Match

```julia
@constraint(model, MS[5, 1] == 1)     # Match 1 takes place at Stadium 5, Estadio Azteca
@constraint(model, MT[1, 3, 1] == 1)  # Match 1 happens on Day 1 at Time 3 (8 PM)
@constraint(model, MC[1, 1] == 1)     # Country 1, México lindo y querido, plays in Match 1
```

### United States Opening Match

```julia
@constraint(model, MS[4, 2] == 1)     # Match 2 is at Stadium 4, SoFi Stadium
@constraint(model, MT[2, 3, 2] == 1)  # Match 2 is on Day 2 at Time 3 (8 PM)
@constraint(model, MC[2, 2] == 1)     # Country 2, America Land of the Free, plays in Match 2
```

### Canada Opening Match

```julia
@constraint(model, MS[12, 3] == 1)    # Match 3 is at Stadium 12, BMO Field
@constraint(model, MT[2, 3, 3] == 1)  # Match 3 is on Day 2 at Time 3 (8 PM)
@constraint(model, MC[3, 3] == 1)     # Country 3, Oh Canada, plays in Match 3
```

## Soccer Tournament Constraints

### Daily Match Limit

```julia
@constraint(model, [d in Days], sum(MD[d, m] for m in Matches) <= LD[d])
```

### One match at a time at any given stadium

```julia
@constraint(model,[d in Days, t in Times, s in Stadiums], 
        sum(MS[s, m] + MT[d, t, m] - 1 for m in Matches) <= 1
    )
```

### 2 Rest days between matches

```julia
@constraint(model, [c in Countries, d in ShortDay], 
        sum(MD[d, m] + MD[d + 1, m] + MD[d + 2, m] - (2 * (1 - MC[c, m]) + 1) for m in Matches) ≤ 0
        
        )
```

### Team plays one match per day at most

```julia
@constraint(model,[c in Countries, d in Days], 
        sum(MC[c, m] + MT[d, t, m] - 1 for t in Times, m in Matches) <= 1
    )
```

## Fairness Constraints

### At least one match per stadium

```julia
@constraint(model, [s in Stadiums], sum(MS[s, m] for m in Matches) >= 1)
```

### No more than one match difference between stadiums

```julia
@constraint(model, [s in 1:(StadiumCount-1)], sum(MS[s, m] for m in Matches) - sum(MS[s+1, m] for m in Matches) <= 1)

@constraint(model, [s in 1:(StadiumCount - 1)], sum(MS[s + 1, m] for m in Matches) - sum(MS[s, m] for m in Matches) <= 1)
```

### Matches are played once

```julia
@constraint(model, [c1 in Countries, c2 in Countries; c1 < c2], 
    sum(MC[c1, m] * MC[c2, m] for m in Matches) <= 1)
```

# Optimization

```julia
optimize!(model)
```

# Results

```julia
using DataFrames

match_schedule = DataFrame(
    Match = Int[],
    Day = Int[],
    Time = Int[],
    Stadium = Int[],
    Country1 = Int[],
    Country2 = Int[],
)

for m in Matches
    day = findfirst(d -> any(t -> value(MT[d, t, m]) > 0.5, Times, Days)
    time = findfirst(t -> any(d -> value(MT[d, t, m]) > 0.5, Days, Times)
    stadium = findfirst(s -> value(MS[s, m]) > 0.5, Stadiums)
    countries = [c for c in Countries if value(MC[c, m]) > 0.5]

    if !isnothing(day) && !isnothing(time) && !isnothing(stadium) && length(countries) == 2
        push!(match_schedule, (
            Match = m,
            Day = day,
            Time = time,
            Stadium = stadium,
            Country1 = countries[1],
            Country2 = countries[2],
        ))
    end
end

# Display the table
match_schedule
```
